```python
# ========================================================
# TASK 2: DIMENSIONALITY REDUCTION USING PCA
# ========================================================

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# 1. Load Dataset
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# 2. Apply PCA to reduce dataset to 2D
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X)

print("PCA applied successfully. Data reduced from 4D to 2D!")

# 3. Find Explained Variance Ratio
print("\n--- PCA EXPLAINED VARIANCE RATIO ---")
print(f"Principal Component 1 (PC1): {pca.explained_variance_ratio_[0]*100:.2f}%")
print(f"Principal Component 2 (PC2): {pca.explained_variance_ratio_[1]*100:.2f}%")
print(
    f"Total Variance Explained: {np.sum(pca.explained_variance_ratio_)*100:.2f}%\n"
)

# 4. Visualize the 2D Transformed Data
plt.figure(figsize=(8, 5))
true_colors = ["teal", "gold", "coral"]
for i in range(3):
    plt.scatter(
        X_pca[y == i, 0],
        X_pca[y == i, 1],
        label=target_names[i],
        c=true_colors[i],
        edgecolor="k",
        s=60,
    )

plt.title("Iris Dataset Projected into 2D Space via PCA")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.grid(True)
plt.show()
```
