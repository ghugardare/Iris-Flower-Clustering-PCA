```python
# ========================================================
# WEEK 4 - TASK 1: MODEL EVALUATION & OPTIMIZATION (SVM)
# ========================================================

import joblib
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.svm import SVC

# 1. Load Dataset
iris = load_iris()
X = iris.data
y = iris.target

# 2. Question 1: Use cross_val_score to test accuracy
svm_model = SVC(kernel="linear", C=1, random_state=42)
scores = cross_val_score(svm_model, X, y, cv=5)  # 5-fold cross validation

print("--- 1. CROSS-VALIDATION SCORES ---")
print(f"Scores for each fold: {scores}")
print(f"Mean CV Accuracy: {scores.mean()*100:.2f}%\n")

# 3. Question 2: Implement GridSearchCV for SVM (Hyperparameter Tuning)
param_grid = {
    "C": [0.1, 1, 10, 100],
    "gamma": [1, 0.1, 0.01, 0.001],
    "kernel": ["rbf", "linear"],
}

grid = GridSearchCV(SVC(), param_grid, refit=True, cv=5)
grid.fit(X, y)

print("--- 2. GRIDSEARCHCV RESULTS ---")
print(f"Best Parameters found: {grid.best_params_}")
print(f"Best Accuracy score: {grid.best_score_*100:.2f}%\n")

# 4. Question 3: Save trained model using joblib
best_svm_model = grid.best_estimator_
joblib.dump(best_svm_model, "best_svm_iris_model.pkl")
print(" Question 3 Success: Model saved as 'best_svm_iris_model.pkl'\n")

# 5. Question 4: Load model and predict on new data
loaded_svm = joblib.load("best_svm_iris_model.pkl")

# New dummy data point (Sepal Length, Sepal Width, Petal Length, Petal Width)
new_data = np.array([[5.1, 3.5, 1.4, 0.2]])
prediction = loaded_svm.predict(new_data)

print("--- 4. MODEL INFERENCE ---")
print(f"Prediction for new data {new_data}: Class {prediction[0]} ({iris.target_names[prediction[0]]})")
```

```python
# ========================================================
# SCRIPT TO EXPORT MODEL DATA TO CSV 
# ========================================================

import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset used in the SVM code
iris = load_iris()

# Create DataFrame
df_iris_svm = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add True Targets and Species Names
df_iris_svm["True_Label"] = iris.target
df_iris_svm["Species_Name"] = [iris.target_names[i] for i in iris.target]

# Save to CSV
df_iris_svm.to_csv("model_optimization_iris_data.csv", index=False)

print("Success: 'model_optimization_iris_data.csv' has been generated on your Jupyter dashboard!")
```
